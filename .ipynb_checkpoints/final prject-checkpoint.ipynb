```python
try:
    import requests
    import json
    y=0
    h=0
    s=0
    n=0
    #lists for the songs 
    songs = []
    happy = []
    sad = []
    neut = []
    
    #the function for determining the mood of the song 
    def ScoreSentiment( positive_text, negative_text, input_text):
        rate = 0
        pos = positive_text.split()
        neg = negative_text.split()
        text = input_text.split()
        for x in text:
            if x in neg:
                rate = rate - 1
            if x in pos:
                rate = rate + 1 
        return (rate)
    
    while True:
    #gets the user input
        song_title = input('Enter the title of the song or enter "quit" to stop: ')
        #this removes the spaces and replaces them wit %20 so the api can understand it
        stitle = song_title.replace(' ', '%20')

        if song_title == 'quit':
            break

        artist = input('Enter the name of the artist: ')
        print('\n')
        #this removes the spaces and replaces them wit %20 so the api can understand it
        art = artist.replace(' ', '%20')

        songs.append({'artist' : artist , 'title' : song_title})
    
        #musixmatch api
        #the api only gives us the first 30% of the lyrics in the song 
        key = "ff46c3844bff96cd0fd10eb67ce06f20"
        url = "https://api.musixmatch.com/ws/1.1/matcher.lyrics.get?format=json&callback=callback&q_track=" + stitle + "&q_artist=" + art + "&apikey="+key
        response = requests.get(url)
        data = response.json()
        input_text = data['message']['body']["lyrics"]['lyrics_body']

        while True:
            #list of words that are positive or negative this determines the mood of the song
            positive_text = ("happy joy yay love enjoy smile yes electric amazing clap believer lucky together excite excitemnt loving wonderful sunshine summertime morning stronger vibe good sensation win winning open right correct")
            negative_text = ("sad angry hate hates dislike pain frown no not cry tears depression depressed anxiety nervous shaking lonely never hurts hurt black mad poison lost lose bleed bleeding loose over fuck tear cheat dump appoligize ugly cut sorry caught close lost dark drown fill void wrong alone")

            ans = ScoreSentiment(positive_text, negative_text, input_text)
            if ans <0: 
                #negative 
                sad.append({'artist' : artist , 'title' : song_title})
            elif ans > 0:
                #positive
                happy.append({'artist' : artist , 'title' : song_title})
            else: 
                #neutral
                neut.append({'artist' : artist , 'title' : song_title})
            break
#print the output
    #gets the number of songs inputed
    for x in songs: 
        y = y+1
    print('\n')
    #prints the songs
    print(y, "Songs inputed:")
    for x in songs: 
        print(x['title'] , 'by' , x['artist'])
        
    print('\n')
    #prints the happy songs
    print("Happy songs:",h)
    for x in happy: 
        print(x['title'] , 'by' , x['artist'])

    print('\n')
    #prints the sad songs
    print("Sad songs:",s)
    for x in sad: 
        print(x['title'] , 'by' , x['artist'])

    print('\n')
    #prints the neutral songs
    print("Neutral songs:",n)
    for x in neut: 
        print(x['title'] , 'by' , x['artist'])
        
except TypeError:
    print('That is not a valid song or artist')

except requests.exceptions.RequestException: 
    print("Cannot connect to the internet") 
    
```
